```python
from scipy.stats import pointbiserialr
from scipy.stats import pearsonr
```

```python
def corr_mean(cat_vars,data = cluster_0):
    corr_dict = {}
    for var in cat_vars:
        corr,p = pointbiserialr(data['rating_median'],data[var])
        mean_T = np.mean(data.loc[data[var] == 1]['rating_median'])
        median_T = np.median(data.loc[data[var] == 1]['rating_median'])
        mean_F = np.mean(data.loc[data[var] == 0]['rating_median'])
        median_F = np.median(data.loc[data[var] == 0]['rating_median'])

        corr_dict[var] = {'corr':round(corr,4) , 'p_val' :p,
                            'mean1': round(mean_T,3) , 'median1': median_T,
                            'mean0': round(mean_F,3) , 'median0': median_F}

    for int_var in ['runtimes', 'votes']:
        corr,p = pointbiserialr(data['rating_median'],data[int_var])
        corr_dict[int_var] = {'corr':round(corr,4) , 'p_val' :p,
                                'mean1': None , 'median1': None,
                                'mean0': None , 'median0': None}
    
    sign_vardf = pd.DataFrame(corr_dict).T
    return(sign_vardf)
```

```python
sign_var_0 = corr_mean(cat_vars = imp_cat_vars,data = cluster_0)
```

```python
sign_var_0.sort_values('p_val')
```
